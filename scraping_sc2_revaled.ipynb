```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
```

```python
def connection():
    YANDEX_DRIVER_PATH = r"C:\yandexdriver\yandexdriver.exe"
    YANDEX_BROWSER_PATH = r"C:\Users\Admin\AppData\Local\Yandex\YandexBrowser\Application\browser.exe"
    
    url = 'https://sc2revealed.com/'

    options = Options()
    options.binary_location = YANDEX_BROWSER_PATH
    options.add_argument("--headless")  # –ë–µ–∑ GUI (–µ—Å–ª–∏ –Ω–µ –Ω—É–∂–µ–Ω –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å)
    options.add_argument("--disable-blink-features=AutomationControlled") # –°–∫—Ä—ã–≤–∞–µ–º selenium-–±–æ—Ç–∞
    
    service = Service(YANDEX_DRIVER_PATH)
    driver = webdriver.Chrome(options=options, service=service)
    driver.get(url)

    return driver
```

```python
def info_GML(driver):
    response = driver.page_source
    
    soup = BeautifulSoup(response)

    rows_players = soup.find_all('div', 'row-player global')

    rows = []
    
    for row_player in rows_players:
        rank = row_player.find('div', 'column rank').find('p', 'player-detail-header').find_next('p').text.strip()[0]
        mmr = row_player.find('div', 'column mmr').find('p', 'player-detail').text
        region = row_player.find('div', 'column region').find('p', 'player-detail').text.strip()
        player = row_player.find('div', 'column player').find('h4').text.split(' ')[1]
        race = row_player.find('div', 'column race').find('p', 'player-detail').text
        record = row_player.find('div', 'column record').find('p', 'player-detail').text.strip()
        last_online = row_player.find('div', 'column last-online').find('p', 'player-detail').text

        try:
            nickname = row_player.find('div', 'column player').find('span').text
        except:
            nickname = '‚Äì'

        rows.append({'Rank': rank, 'MMR': int(mmr), 'Region': str(region), 'Player': player, 'Nickname': nickname, 'Race': race, 'Record': record, 'Last online': last_online})

    df = pd.DataFrame(data=rows)
    return df
            
        # print(rank)
        # print(player)
        # print(nickname)
        # print(race)
        # print(record)
        # print(last_online)
        # print(region)
```

```python
%%time
driver = connection()
```

```python
%%time
df = info_GML(driver)
df.head()
```

## –î–æ–±–∞–≤–∏–º –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞ Wins –∏ Losses

```python
df['Wins'] = df['Record'].apply(lambda x: int(x.split()[0]))
df['Losses'] = df['Record'].apply(lambda x: int(x.split()[2]))
```

```python
df.head()
```

```python
df['Win_rate'] = round(df['Wins'] / (df['Wins'] + df['Losses']), 2) * 100
df.head()
```

```python
df[df['Region'] == 'EU'].groupby('Race').agg(Winrate=('Win_rate', 'mean'), AVG_MMR=('MMR', 'mean'), Amount=('Race', 'count'))
```
